# TOKENIZADOR DE LA BIBLIOTECA DE REPROCESAMIENTO DE KERAS


```python
from tensorflow.keras.preprocessing.text import Tokenizer
```

```python
sentences = [
    'i love my dog',
    'I, love my cat',
    'You love my dog'
]

tokenizer = Tokenizer(num_words = 100)
tokenizer.fit_on_texts(sentences)
word_index = tokenizer.word_index
print(word_index)
```

# TEXTO A SECUENCIA


```python

sequences = tokenizer.texts_to_sequences(sentences)
print(word_index)
print(sequences)
```

# Probando con otras frases

```python
#Formando un CORPUS
oraciones = [
    'El celular es de color negro',
    'El color negro es espectacular',
    'Mi espectacular celular negro'
]

tokenizer = Tokenizer(num_words = 100)
tokenizer.fit_on_texts(oraciones)
word_index = tokenizer.word_index
print(word_index)
```

```python
#Formando la lista de valores basada en los tokens
sequences = tokenizer.texts_to_sequences(oraciones)
print(word_index)
print(sequences)
```
